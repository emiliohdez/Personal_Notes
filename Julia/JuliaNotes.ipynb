```julia
# Declares a function f(x, y) with the "assignment form"
f(x, y) = x^2 + y^2

# Declares two more functions above the f(x, y).
f0(x) = f(x, 0)
f2(x) = f(x, 2)

# Declares two ranges for x and y, with limits in -5 and 5, and steps of 0.1.
xVals, yVals = -5:0.1:5, -5:0.1:5


plot(xVals, [f0.(xVals), f2.(xVals)], c = [:blue :green :red],
    xlims = (-5, 5), legend = :top, ylims = (-5, 25),
    ylabel = L"f(x, \cdot)", xlabel = "label for x-axis", label = [L"f(x, 0)" L"f(x, 2)"])

#=
    Attributes for this plot():
    xlabel => is the label for the x-axis.
    ylabel => is the label for the y-axis.
    xlims => the limits for the x dimension.
    ylims => the limits for the y dimension.
    c => describes the colors that can be used for plotting.
    label => is the label for the color keys.
    legend => specifies the position of the label for color keys.
=#

p1 = annotate!(0, -0.5, text("(0,0) The minimum\n of f(x,0)", :left, :top, 10));
plot(p1)
```

```julia
z = [ f(x, y) for y in yVals, x in xVals]
p2 = surface(xVals, yVals, z, c = cgrad([:yellow, :green]), legend = :none,
    ylabel = "y", zlabel = L"f(x,y)")
plot(p2, size = (600, 600))
```

```julia
M = z[1:10, 1:10]
p3 = heatmap(M, c = cgrad([:blue, :red]), yflip = true, ylabel = "y",
    xticks = ([1:10;], xVals), yticks = ([1:10;], yVals))
plot(p3)
```

```julia
function hailLength(x::Int)
   n = 0
    while x != 1
        if x % 2 == 0
            x = Int(x/2)
        else
            x = 3x + 1
        end
            x += 1
    end
    return n
end
    
lengths = [hailLength(x0) for x0 in 2:10^7]
histo = histogram(lengths, bins = 1000, normed = :true, fill = (:blue, true), la = 0,
        legend = :none, xlims = (0, 500), ylims = (0, 0.012), xlabel = "Length",
        ylabel = "Frequency")
plot(histo)
```
